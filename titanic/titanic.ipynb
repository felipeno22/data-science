#***Gráfico***


*   usando lib plotly
*   histograma para  mostras os sobreviventes mulheres e homensn



```python
import plotly.express as px


# Gráfico Bar conta o total de passageiros
total_passageiros = len(dataset)#total pass
print(total_passageiros)
# Criar o gráfico de barras para o total de passageiros
grafico_total_pass = px.bar(
    x=["Total de Passageiros"],  # Rótulo do eixo X
    y=[total_passageiros],        # Valor no eixo Y
    text=[total_passageiros],     # Mostrar o valor na barra
    title="Total de Passageiros",
    labels={"x": "Categoria", "y": "Quantidade"}
)
# Exibir o gráfico
grafico_total_pass.show()

# Gráfico Bar Conta o total de passageiros com idade nula
#conta todos que tem idade nula
total_pass_age_null=dataset['Age'].isnull().sum()
print(total_pass_age_null)
grafico_total_pass_age_null = px.bar(
    x=["Total de Passageiros Age Null"],  # Rótulo do eixo X
    y=[total_pass_age_null],        # Valor no eixo Y
    text=[total_pass_age_null],     # Mostrar o valor na barra
    title="Total de Passageiros Age Null",
    labels={"x": "Categoria", "y": "Quantidade"}
)
# Exibir o gráfico
grafico_total_pass_age_null.show()

# Gráfico Bar Conta o total de passageiros com idade nula
#conta todos que tem idade nula
total_pass_age=dataset['Age'].notnull().sum()
print(total_pass_age)
grafico_total_pass_age = px.bar(
    x=["Total de Passageiros Age"],  # Rótulo do eixo X
    y=[total_pass_age],        # Valor no eixo Y
    text=[total_pass_age],     # Mostrar o valor na barra
    title="Total de Passageiros Age ",
    labels={"x": "Categoria", "y": "Quantidade"}
)
# Exibir o gráfico
grafico_total_pass_age.show()

#grafico de quantos sobreviveram e morreram por sexo
grafico_survived_sex = px.histogram(dataset, x='Sex', color="Survived", text_auto=True)
#exibir o grafico
grafico_survived_sex.show()

#grafico de quantos sobreviveram e morreram por CLASS
grafico_survived_pclass = px.histogram(dataset, x='Pclass', color="Survived", text_auto=True)
#exibir o grafico
grafico_survived_pclass.show()

# Gráfico Bar Conta o total de passageiros que sobreviveram por sexo
total_pass_surviver_sex=dataset[dataset["Survived"]==1].groupby(['Sex']).size()
print(total_pass_surviver_sex)
# Gráfico Bar Conta o total de passageiros sobrevivente por sexo
grafico_total_pass_survived_sex = px.bar(
    total_pass_surviver_sex.reset_index(),
    x="Sex",  # Rótulo do eixo X
    y=0,        # Valor no eixo Y
    text=0,     # Mostrar o valor na barra
    title='Total de Passageiros Sobreviventes do Sexo Feminino',
    labels={'Sex': 'Sexo', '0': 'Quantidade'}
)
# Exibir o gráfico
grafico_total_pass_survived_sex.show()


# Gráfico Bar Conta o total de passageiros que sobreviveram por classe
total_pass_surviver_classe=dataset[dataset["Survived"]==1].groupby(['Pclass']).size()
print(total_pass_surviver_classe)
# Gráfico Bar Conta o total de passageiros sobrevivente por classe
grafico_total_pass_survived_classe = px.bar(
    total_pass_surviver_classe.reset_index(),
    x="Pclass",  # Rótulo do eixo X
    y=0,        # Valor no eixo Y
    text=0,     # Mostrar o valor na barra
    title='Total de Passageiros Sobreviventes do Sexo Classe',
    labels={'Pclass': 'Classe', '0': 'Quantidade'}
)
# Exibir o gráfico
grafico_total_pass_survived_classe.show()

#grafico que calcula a media de idade dos pass com idade
# Gráfico Bar Conta o total de passageiros sobrevivente por classe
media_age=float(f" {media_age:.2f}")
grafico_media_age_pass = px.bar(
    x=["Média de idade"],  # Rótulo do eixo X
    y=[media_age],        # Valor no eixo Y
    text=[media_age],     # Mostrar o valor na barra
    title='Média de idade Passageiros',
     labels={"x": "Idade", "y": "Média"}
)
# Exibir o gráfico
grafico_media_age_pass .show()


```
